```python
#Imports
import pdfplumber
from docx import Document
```

```python
fp=r'C:\Users\Andrew Chan\OneDrive\Documents\中醫食療概念\課程筆記_880168_中醫食療概念_Internet Mode_PW.pdf'
change_dot=fp.replace('.','\\')
fp_split=change_dot.split("\\")
doc_name=fp_split[-2]
```

```python
#Get all text
doc = Document()
with pdfplumber.open(fp) as pdf:
    pages=pdf.pages
    for i,pg in enumerate(pages):
        text = pages[i].extract_text()
        print(text)
        content = doc.add_paragraph(text)
```

```python
#Convert to document
file_name=doc_name+'.docx'
doc.save(file_name)
```
